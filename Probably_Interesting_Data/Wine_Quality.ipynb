```python
def render (features, y):
    plt.figure (figsize = (40, 30))
    for x in features:
        plt.subplots ()
        plt.title(y + ' by ' + x)
        plt.xlabel(x)
        plt.ylabel(y)
        plt.scatter (wine_data [x], wine_data [y])
    plt.show ()
```

```python
predict_target = 'fixed acidity'
wine_features = ['citric acid', 'pH','alcohol']
render (wine_features, predict_target)
```

```python
# Choose to model the relationship between pH and fixed acidity since they seems to be nicely distributed
fixed_acidity = wine_data ['fixed acidity'].values
pH = wine_data ['pH'].values

# Each point dictionary contains of x: fixed_acidity, y: pH
# Return a list of points dictionaries 
def get_point_list (dataset):
    points = []
    for i in range(0, len(dataset['0'])):
        point = {}
        point['x'] = dataset['0'][i]
        point['y'] = dataset['1'][i]
        points.append(point)
    return points

points = get_point_list ({'0': pH, '1': fixed_acidity})
```

```python
clusters, centers = Kmeans(points, 3)
```
