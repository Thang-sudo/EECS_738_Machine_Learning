```python
# The mixture model that I use in this project is K-means clustering.
# I choose this Iris species dataset since K-means clustering can not deal with categorical data, and this dataset provide numerical data only
import pandas as pd
import import_ipynb
import matplotlib.pyplot as plt
from Kmeans import *
%matplotlib inline
```

```python
iris_data = pd.read_csv ('./Iris.csv')
```

```python
def render (features, y):
    plt.figure (figsize = (40, 30))
    for x in features:
        plt.subplots ()
        plt.title(y + ' by ' + x)
        plt.xlabel(x)
        plt.ylabel(y)
        plt.scatter (iris_data [x], iris_data [y])
    plt.show ()
```

```python
# I choose the predict target to be SepalLengthCm
# Other attributes of the Iris flower will be the features
predict_target = 'SepalLengthCm'
iris_features = ['SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']

# Render the distributions of the SepalLength correspond to other features
render (iris_features, predict_target)
```

```python
sepal_length = iris_data ['SepalLengthCm'].values
petal_length = iris_data ['PetalLengthCm'].values

# Each point dictionary contains of x: petalLength, y: sepalLength
# Return a list of points dictionaries 
def get_point_list (dataset):
    points = []
    for i in range(0, len(dataset['0'])):
        point = {}
        point['x'] = dataset['0'][i]
        point['y'] = dataset['1'][i]
        points.append(point)
    return points

points = get_point_list ({'0': petal_length, '1': sepal_length})
```

```python
clusters, centers = Kmeans(points, 4)
```
